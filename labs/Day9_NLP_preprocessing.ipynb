```python
import pandas as pd

ts_lyrics = pd.read_csv('/content/drive/MyDrive/taylor_swift_lyrics.csv')
```

```python
ts_lyrics.head()
```

```python
import re

def remove_linebreaks(text):
  return re.sub('\n', ' ', text)

ts_lyrics['processed_lyrics'] = ts_lyrics['Lyrics'].apply(lambda text:remove_linebreaks(text))
```

```python
ts_lyrics.head()
```

```python
def remove_punctuation(text):
  return text.translate(str.maketrans('', '', string.punctuation))

ts_lyrics['processed_lyrics'] = ts_lyrics['processed_lyrics'].apply(lambda text:remove_punctuation(text))
```

```python
ts_lyrics.head()
```

```python
ts_lyrics['processed_lyrics'] = ts_lyrics['processed_lyrics'].str.lower()
```

```python
from collections import Counter

text = "".join(lyric for lyric in ts_lyrics['processed_lyrics'])

split_it = text.split()
count = Counter(split_it)
most_occur = Counter.most_common(count)
print(most_occur)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
stop_words.update(['im', 'youre', 'id', 'dont', 'cant', 'didnt', 'ive', 'ill', 'hasnt'])
print(stop_words)
```

```python
def remove_stopwords(text):
  return " ".join([word for word in str(text).split() if word not in stop_words])

ts_lyrics['processed_lyrics'] = ts_lyrics['processed_lyrics'].apply(lambda text:remove_stopwords(text))
```
